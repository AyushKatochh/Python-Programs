```python
import matplotlib.pyplot as plt

# x axis values
x = [5,2,9,4,7]

# y values
y = [10, 5, 8, 4, 2]

# Function to plot
plt.plot(x, y)

plt.title("Amplitude vs Time")
plt.ylabel("Amplitude")
plt.xlabel("Time")

# Function to show
plt.show()
```

```python
month = ["Jan", "Feb", "March", "April", "May"]
SBI = [515.00, 502.70, 513.00, 512.00, 510.95]
PNB = [42.00, 41.55, 42.25, 42.20, 41.80]

plt.plot(month, SBI, "b", label="SBI", linewidth=2)
plt.plot(month, PNB, "r", label="PNB", linewidth=2)

plt.title("Top Banking Sector in India by Market Corporation")
plt.ylabel("Share Price")
plt.xlabel("Month")
plt.legend()
plt.show()
```

```python
from deepface import DeepFace
# Load and display input image
import cv2
import matplotlib.pyplot as plt
img1 = cv2.imread("/content/image1.jpg")
plt.imshow(img1)
plt.show()
```

```python
# plotting the image 
# Flipping the color from BGR to RGB
plt.imshow(img1[:,:, ::-1])
plt.show()
```

```python
#Analyse Image
obj = DeepFace.analyze(img1, actions=['age', 'gender', 'race', 'emotion'])
print(obj)
```

```python
#print result
print("Your age is", obj["age"],
      "\nand you are", obj["dominant_race"],
      "\nand current emotion is", obj["dominant_emotion"],
      "and your gender is", obj["gender"])
```
